```python
try:
    raw_hour = (input("Insert your weekly worked hours: ").strip())
    hour = int(raw_hour)

except ValueError:
    print(f"Error: you typed {raw_hour}, please insert a number")
    raise SystemExit

try:
    raw_rate = (input("Insert your rate: ").strip().replace(",", "."))
    rate = float(raw_rate)

except:
    print(f"Error: you typed {raw_rate}, please insert a number")
    raise SystemExit

def computepay (hr, rt):
    if hr > 40:
        normal_h = hr - 40
        total_h = (rt * 40) + (rate * normal_h * 1.5)
        print(f"Your weekly worked hours are: {hr}. Your weekly payment is: {total_h:.2f}")

    else:
        regular_p = rt * hr
        print(f"Your weekly worked hours are: {hr}. Your weekly payment is: {regular_p}")

computepay(hour, rate)

```

```python


subjects = ["Math", "English", "History"]
note = 0

for list_subjects in subjects:
    add_note = float(input(f"Insert your {list_subjects} note: ").strip().replace(",", "."))
    note += add_note

final_note = note / len(subjects)

def final_notes(final_note):
    rules = {
     #    key            value
    lambda n: n >= 9: "A - Approved ✅",
    lambda n: n >= 8: "A - Approved ✅",
    lambda n: n >= 7: "C - Educational Recovery ⚠️",
    lambda n: n >= 6: "D - Failed ❌",
    lambda n: n < 6 :  "F - Failed ❌"
    }   # The lambda is a function, but inside the dictionary, it’s used as a key.

    for notes, result in rules.items():
        # notes - variable that holds the functions from the key
        # result - variable that holds the data from the value
        if notes(final_note):
        # notes here is a variable that holds a function (one of the lambdas).
        # You are executing the function stored in notes, passing the value of final_note as input.
            return f"Your final note is {final_note:.2f} → {result}"

print(final_notes(final_note))
 
```
