```python
try:
  
  play = int(input("Insira um número entre 10 e 20: ").strip())

  if play< 10:
    print("Número inválido, menor que 10!")
    
  elif play > 20:
    print("Número inválido, maior que 20!")

  else:
    print(f"Você digitou: {play}")
    
except ValueError:
    print("Você não digitou um número!")
  
```

```python
number = [3,41, 12, 9, 74, 15]

for numbers in number:
    if numbers != 74:
        print (f"{numbers} it's not the biggest number")

    else: 
         print(f"✅{numbers} IT's THE BIGGEST NUMBER✅")
```

```python
largest = -1
number = 1

print("Before", largest)

for the_lists in [2, 65, 41, 89, 32, 78, 42, 26300, 236, 25 ,1]:
    if the_lists > largest:
        largest = the_lists
        
    print(number, the_lists)
    number += 1
        
    
print("After", largest)

 
    
  

```
