```python
real_1 = int(input("Enter how many 1 real coins you have: "))
cent_50 = int(input("Enter how many 50 cent coins you have: "))
cent_25 = int(input("Enter how many 25 cent coins you have: "))
cent_10 = int(input("Enter how many 10 cent coins you have: "))
cent_5 = int(input("Enter how many 5 cent coins you have: "))

total_value = (
    (real_1 * 1.00)
    + (cent_50 * 0.50)
    + (cent_25 * 0.25)
    + (cent_10 * 0.10)
    + (cent_5 * 0.05)
)

print(f"You have a total of R${total_value:.2f}.")
```

```python
total_value = 0

coins = [
    ("1 real", 1.00),
    ("50 centavos", 0.50),
    ("25 centavos", 0.25),
    ("10 centavos", 0.10),
    ("05 centavos", 0.05)
]

for name, value in coins:
    quantity = int(input(f"Enter how many {name} you have: "))
    total_value += quantity * value 


print(f"You have a total of R${total_value:.2f}.")
```

```python


final_notes = 0


informations = ["Math", "English", "History"]

for note_info in informations:
    answer = (input(f"Insert your final {note_info} note: ").strip().replace(",", "."))
    answer_f = float(answer)
    final_notes += answer_f

    
# Atenção ao recuo do print ou similares, se estiver dentro do loop 
# vai imprimir os resultados individualmente.

final_notes/= 3

print(f"Your final note is: {final_notes:.2f}")

if final_notes >= 6:
    print("Approved")
   
elif final_notes >= 4 and final_notes <= 5.9:
    print("Educational Recovery")

else:
    print("Failed")
```

```python
subjects = ["Math", "English", "History", "Geo"]
score = 0

for sub_info in subjects:
    answer = float(input(f"Insert your final {sub_info} score:").strip().replace(",", "."))
    score += answer

score /= len(subjects)
# len() return the numbers of items in a list

print(f"Your final score is: {score:.2f}")

if score >= 6:
    print("✅Approved")

elif score >= 4 and score <=5.9:
    print("⚠️Educational Recovery")

else: 
    print("❌Failed")

# Watchout the Indentation in Python:
# for i in range(5):         start of the loop
#   print(i)               ← inside the for loop
#   if i > 2:              condition inside the for loop
#      print('Bigger than 2')  ← inside the if statement
#  print('Done with i', i)     ← still inside the for loop
#print('All Done')          ← outside the for loop
```

```python
hours = int(input("Enter your week hours worked: ").strip())
rate = float(input("Enter your rate:  ").strip().replace(",", "."))

if hours > 40:
    weekly_hours = 40
    extra_hours  = hours - weekly_hours
    hours_extra  = extra_hours * rate * 1.5
    total_hours = weekly_hours * rate + hours_extra
    print(f"Your hours is {hours}. Your weeekly payment is {total_hours:.2f}")

else:
    total = rate * hours
    print(f"Your hours is {hours}. Your weeekly payment is {total:.2f}")
```

```python

try:
    hours = (input("Enter your week hours worked: ").strip())
    hours_i = int(hours)

except ValueError:
    print(f"Error: you typed '{hours}' please type a number.")
    raise SystemExit

try: 
    rate = (input("Enter your rate: ").strip().replace(",", "."))
    rate_f = float(rate)

except ValueError:
     print(f"Error: you typed '{rate}' please type a number")

else:
    if hours_i > 40:
        hours_e = hours_i - 40
        payment_e = (rate_f * 40) + (rate_f * hours_e * 1.5)
        print(f"Your total weekly hour is: {hours_i}. Your payment will be: {payment_e:.2f}")

    else:
        payment_n = rate_f * hours_i
        print(f"Your total weekly hours is: {hours_i}. Your paymente will bie: {payment_n:.2f}")
```
