```python
real_1 = int(input("Enter how many 1 real coins you have: "))
cent_50 = int(input("Enter how many 50 cent coins you have: "))
cent_25 = int(input("Enter how many 25 cent coins you have: "))
cent_10 = int(input("Enter how many 10 cent coins you have: "))
cent_5 = int(input("Enter how many 5 cent coins you have: "))

total_value = (
    (real_1 * 1.00)
    + (cent_50 * 0.50)
    + (cent_25 * 0.25)
    + (cent_10 * 0.10)
    + (cent_5 * 0.05)
)

print(f"You have a total of R${total_value:.2f}.")
```

```python
total_value = 0

coins = [
    ("1 real", 1.00),
    ("50 centavos", 0.50),
    ("25 centavos", 0.25),
    ("10 centavos", 0.10),
    ("05 centavos", 0.05)
]

for name, value in coins:
    quantity = int(input(f"Enter how many {name} you have: "))
    total_value += quantity * value 


print(f"You have a total of R${total_value:.2f}.")
```

```python
hours = int(input("Enter your week hours worked: ").strip())
rate = float(input("Enter your rate:  ").strip().replace(",", "."))

if hours > 40:
    weekly_hours = 40
    extra_hours  = hours - weekly_hours
    hours_extra  = extra_hours * rate * 1.5
    total_hours = weekly_hours * rate + hours_extra
    print(f"Your hours is {hours}. Your weeekly payment is {total_hours:.2f}")

else:
    total = rate * hours
    print(f"Your hours is {hours}. Your weeekly payment is {total:.2f}")
```

```python

try:
    hours = (input("Enter your week hours worked: ").strip())
    hours_i = int(hours)

except ValueError:
    print(f"Error: you typed '{hours}' please type a number.")
    raise SystemExit

try: 
    rate = (input("Enter your rate: ").strip().replace(",", "."))
    rate_f = float(rate)

except ValueError:
     print(f"Error: you typed '{rate}' please type a number")

else:
    if hours_i > 40:
        hours_e = hours_i - 40
        payment_e = (rate_f * 40) + (rate_f * hours_e * 1.5)
        print(f"Your total weekly hour is: {hours_i}. Your payment will be: {payment_e:.2f}")

    else:
        payment_n = rate_f * hours_i
        print(f"Your total weekly hours is: {hours_i}. Your paymente will be: {payment_n:.2f}")
```

```python
subjects = ["Math", "English", "History"]
note = 0

for list_subjects in subjects:
    add_note = float(input(f"Insert your {list_subjects} note: ").strip().replace(",", "."))
    note += add_note

final_note = note / len(subjects)

def final_notes(final_note):
    rules = {
     #    key            value
    lambda n: n >= 9: "A - Approved ✅",
    lambda n: n >= 8: "A - Approved ✅",
    lambda n: n >= 7: "C - Educational Recovery ⚠️",
    lambda n: n >= 6: "D - Failed ❌",
    lambda n: n < 6 :  "F - Failed ❌"
    }   # The lambda is a function, but inside the dictionary, it’s used as a key.

    for notes, result in rules.items():
        # notes - variable that holds the functions from the key
        # result - variable that holds the data from the value
        if notes(final_note):
        # notes here is a variable that holds a function (one of the lambdas).
        # You are executing the function stored in notes, passing the value of final_note as input.
            return f"Your final note is {final_note:.2f} → {result}"

print(final_notes(final_note))
```

```python
from unidecode import unidecode 
import string

original_phrase = input("Insert a random word or phrase: ")
espacos = original_phrase.replace(" ", "")
acentos = unidecode(espacos)
# "unidecode" the library transliterates accented characters to their unaccented equivalents — that is, ‘á’ becomes ‘a’.”
lower_word = acentos.lower()

remover_pont_tabela = lower_word.translate(str.maketrans('', '', string.punctuation))
# str.maketrans" creates a translation table that ".translate" will use to perform conversions based on that table. 
# Since I didn’t define any parameters in "str.maketrans", leaving the table blank with '', '',, the ".translate" 
# method understands that everything contained in "string.punctuation" (a string imported that includes all punctuation characters)
# should be replaced or removed.
inv = remover_pont_tabela[::-1]


if inv.isalpha():

    if inv == remover_pont_tabela:
    
     print(f"Sua palavra é um palíndromo: original: {original_phrase}. Sua frase ou palavra é um palíndromo: {inv}")

    elif inv != remover_pont_tabela:
       
     print(f"Sua palavra é um palíndromo: original: {remover_pont_tabela}, não é um palíndromo: {inv}")
else:
    print("Try again, type a word or phrase")
```
